# ¿Qué son los Transformers?

Los transformers es la base lógica y arquitectura neunrolan detrás de los LLM's

Es el marco dentro del cual se crea el modelo de lenguaje.


Transformadores de manera general, estos pueden agruparse en tres categorías:

* Parecidos a GPT (también llamados modelos auto-regressive)
* Parecidos a BERT (también llamados modelos auto-encoding)
* Parecidos a BART/T5 (también llamados modelos sequence-to-sequence)

Todos los modelos de Transformadores mencionados con anterioridad han sido entrenados como modelos de lenguaje. Esto significa que han sido entrenados con grandes cantidades de texto crudo de una manera **auto-supervisada.**

El aprendizaje auto-supervisado es un tipo de entrenamiento en el que el objetivo se computa automáticamente de las entradas del modelo. ¡Esto significa que no necesitan humanos que etiqueten los datos!

## Cómo es que llegamos a un LLM con ayuda de la arquitectura de los transformers?

Los transformes es la base logica que tendremos para que podamos preentrenar un modelo atraves del aprendizaje autosupervisado, una vez que pasa a eso, le hacemos transfer learning a traves de la técnica fine-tunning usando aprendizaje supervisado, lo que implica que se usaron etiquetas hechas por humanos para una tarea específica, posterior al transfer learning este se convierta en un llm.

Un ejemplo de una tarea es predecir la palabra siguiente en una oración con base en las n palabras previas. Esto se denomina modelado de lenguaje causal porque la salida depende de las entradas pasadas y presentes, pero no en las futuras.

My name is -> [PREDICCIÓN] = Edu

Otro ejemplo es el modelado de lenguaje oculto, en el que el modelo predice una palabra oculta en la oración.

My [PREDICCIÓN] is Edu

## Encoder - Decoder

image.png

Pueden ser usador en conjunto, pero también pueden ser usados de manera independiente

### Encoder

1. El encoder acepta el input (texto)
2. Convierte este texto en una representación numérica **(embedding)**
3. Estos embeddings son usados por el mecanismo de atención (self-attention) como su componente principal
4. Una vez usado por este mecanismo son enviados al decoder

### Decoder

1. también puede aceptar los mismos inputs como el encoder
2. Convierte este texto en una representación numérica **(embedding)**
3. El encoder es también la autoatención enmascarada.
4. Se diferencia del "encoder" debido a su propiedad unidireccional y se utiliza tradicionalmente de manera autorregresiva.

La combinación del encoder y decoder, es conocida como **"encoder-decoder"** o **"sequece-tosequence"**

* Encoder acepta inputs y representa estos inputs en embbedings

* Estos embeddings son usados por el mecanismo de atención como su componente principal y lanzando un output que serán vectores de representación contextualizados

* Estos vectores resultantes del proceso del self-attention son pasados al decoder

* Decoder usa los outputs del encoder y junto con los inputs que tiene el decoder los usa para generar una predicción , la cual se transforma en un ouput

* El output que se obtiene será reusado en el proceso "auto-regressive"

* Una vez echo esto , significa que el modelo está optimizado para generar salidas con respecto al primer input que llegó en el encoder

image.png

Cada una de estas partes puede ser usada de manera independiente, dependiendo de la tarea:

* Modelos con solo codificadores: Buenos para las tareas que requieren el entendimiento de la entrada, como la clasificación de oraciones y reconocimiento de entidades nombradas.
* Modelos con solo decodificadores: Buenos para tareas generativas como la generación de textos.
* Modelos con codificadores y decodificadores o Modelos secuencia a secuencia: Buenos para tareas generativas que requieren una entrada, como la traducción o resumen.

# Capas de Atención

Esta capa va a indicarle al modelo que tiene que prestar especial atención a ciertas partes de la oración que le pasaste (y más o menos ignorar las demás), cuando trabaje con la representación de cada palabra.

Piensa en la tarea de traducir texto de Inglés a Francés. Dada la entrada “You like this course”, un modelo de traducción necesitará tener en cuenta la palabra adyacente “You” para obtener la traducción correcta de la palabra “like”, porque en Francés el verbo “like” se conjuga de manera distinta dependiendo del sujeto. Sin embargo, el resto de la oración no es útil para la traducción de esa palabra. En la misma línea, al traducir “this”, el modelo también deberá prestar atención a la palabra “course”, porque “this” se traduce de manera distinta dependiendo de si el nombre asociado es masculino o femenino. De nuevo, las otras palabras en la oración no van a importar para la traducción de “this”. Con oraciones (y reglas gramaticales) más complejas, el modelo deberá prestar especial atención a palabras que pueden aparecer más lejos en la oración para traducir correctamente cada palabra.

El mismo concepto aplica para cualquier tarea asociada con lenguaje natural: una palabra por si misma tiene un significado, pero ese significado está afectado profundamente por el contexto, que puede ser cualquier palabra (o palabras) antes o después de la palabra que está siendo estudiada.
