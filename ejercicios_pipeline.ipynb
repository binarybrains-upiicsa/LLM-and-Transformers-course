## Función Pipeline

En la librería **Transformers de Hugging Face**, la función pipeline es una API de muy alto nivel diseñada para simplificar al máximo el uso de modelos preentrenados en tareas comunes de NLP (Procesamiento de Lenguaje Natural). En lugar de tener que ocuparte manualmente de cargar el tokenizador, el modelo, convertir tus textos a tensores, ejecutar la inferencia y luego post-procesar la salida, con pipeline haces todo eso en un solo paso.

```python
import transformers
from transformers import pipeline

analisis_sentimiento = pipeline("sentiment-analysis")
analisis_sentimiento(
    ["This movie has made me inspired by emotion"
    ,"Picture have a special has a touch of emotion"])
```

```python
from transformers import pipeline

generator = pipeline("text-generation", model="distilgpt2")
generator(
    "Este curso te va a ayudar a",
    max_length=30,
    num_return_sequences=2,
)
```
